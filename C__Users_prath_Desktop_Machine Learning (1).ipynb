# Importing the Libraies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading the data

```python
header = ["variance", "skew", "curtsies", "entropy", "classes"]
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt"
df = pd.read_csv(url, names=header)
```

```python
df.head()
```

# Performing descriptive analysis and determining if the data contain null values. 

# Is null() command is used to check if the give data has any NULL values, If will return if the given data any null values or else it will return false.

```python
df.isnull()
```

# isnull.sum() will give the column-wise sum of missing values

```python
df.isnull().sum()
```

# Visualization of data using seaborn library(sns) pairplot method

```python
g = sns.pairplot(df, hue='classes', markers='+')
plt.show()
```

```python
df.hist(edgecolor = 'white', figsize = (20,16))
```

# Preparing the train set and the test set with 40% as test size and randon_state=1 to see constant output,and Y axis as classes.

```python
X = df.drop(['classes'], axis=1)
y = df['classes']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import Perceptron

ppn = Perceptron(max_iter=40, eta0=0.1, random_state=1)
ppn.fit(X_train, y_train)
```

#  Getting the prediction value, Calculating the accuracy of the model

```python
y_pred = ppn.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```
